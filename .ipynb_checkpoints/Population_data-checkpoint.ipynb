```python
# importing dependencies
import pandas as pd
```

```python
# establishing path and reading Imported Deforestation data into pandas
population_df = pd.read_csv(r"C:/Users/abrow/Desktop/EMERSON_BOOTCAMP/BOOTCAMP_AMB/Project_3/Project-3-Global-Warming/Data/Deforestation/owid-co2-data.csv")
population_df.head()
```

```python
#Extracting the desired columns and creating a new dataframe
jennycraig_pop_df = population_df[['iso_code', 'country', 'year', 'population']].copy()
jennycraig_pop_df.head()
```

```python
# Renaming the columns to be compatible with the other dataframes that it will be merged with
jennycraig_pop_df.columns = ['Code', 'Entity', 'Year', 'Population']
jennycraig_pop_df.head()
```

```python
# Checking data types to ensure compatability for ease of merging
jennycraig_pop_df.dtypes
```

```python
# Well I don't feel like counting how many different codes there are, so I'll make Pandas count the number of unique values for me...
# Creating a variable to hold the count
count = 0

# Creating an empty list to hold the counted values
counted = []

# loop for counting the unique values in in the code column
for i in range(0, len(jennycraig_pop_df['Code'])):
    
    if jennycraig_pop_df['Code'][i] not in counted:
        
        counted.append(jennycraig_pop_df['Code'][i])
        
        count += 1
        
    print(count)
```

```python
# Dropping rows where a country's population value is Nan
slimfast_pop_df = w8watchers_pop_df[w8watchers_pop_df["Population"].notna()]
```

```python
# Reordering the columns so that I can merge the dataframes smoothly
slimfast2_df = slimfast_pop_df[['Entity', 'Code', 'Year', 'Population']]
slimfast2_df.head()
```

```python
# Creating path to read in Net Deforestation data
netDeforestation_df = pd.read_csv(r"C:\Users\abrow\Desktop\EMERSON_BOOTCAMP\BOOTCAMP_AMB\Project_3\Project-3-Global-Warming\Data\Deforestation\readyToMerge_NetDeforestation.csv")
netDeforestation_df.head()
```

```python
# Creating path to read in Imported Deforestation data
importedDeforestation_df = pd.read_csv(r"C:\Users\abrow\Desktop\EMERSON_BOOTCAMP\BOOTCAMP_AMB\Project_3\Project-3-Global-Warming\Data\Deforestation\readyToMerge_ImportedDeforestation.csv")
importedDeforestation_df.head()
```

```python
#Well now we have 2 indexes that we don't need, but I guess that's not a big deal
# Comparing the "Code" series in Net and Imported deforestation dataframes first
netDeforestation_df['Code'].isin(importedDeforestation_df['Code']).value_counts()
```

```python
#Okay, awesome, they all match
# Creating a variable for the extracted column
extracted_col = netDeforestation_df["Net Traded Deforestation"]
```

```python
# Extracting the Net Deforestation column from NetDeforestation_df and adding to ImportedDeforestation_df
Deforestation_df = importedDeforestation_df.join(extracted_col)
print(Deforestation_df)
```

```python
# Deleting the Unamed: 0 Column because we don't need it
Deforestation_df =Deforestation_df.drop('Unnamed: 0', 1)
```

```python
# Okay, that's much better
# Now to compare the population dataframe to the Deforestation dataframe
Combined_df = slimfast2_df.loc[slimfast2_df['Code'].isin(Deforestation_df['Code'])]
print(Combined_df)
```

```python
#Well I'm missing my other 2 columns, so that's no bueno
#Flip-flopping them to see if the order I did it in matters
Combined2_df = Deforestation_df.loc[Deforestation_df['Code'].isin(slimfast2_df['Code'])]
print(Combined2_df)
```

```python
# Apparently it did matter, now to merge them
merged_df = pd.merge(Deforestation_df, slimfast2_df, on=['Entity', 'Code', 'Year'], how='left')
print(merged_df)
```

```python
# Making it's display more presentable
merged_df.head()
```

```python
#Calculating per Capita Imported Deforestation
merged_df['per Capita Imported Deforestation'] = merged_df['Imported Deforestation']/merged_df['Population']
merged_df.head()
```

```python
# Calculating per Capita Net Traded Deforestation
merged_df['per Capita Net Traded Deforestation'] = merged_df['Net Traded Deforestation']/merged_df['Population']
merged_df.head()
```

```python
# Exporting cleaned data to a CSV 
# Specifying a path to a folder for cleaned data
merged_df.to_csv(r'C:/Users/abrow/Desktop/EMERSON_BOOTCAMP/BOOTCAMP_AMB/Project_3/Project-3-Global-Warming/Data/Deforestation/Clean_data/clean_deforestation.csv', encoding='utf-8')
```
